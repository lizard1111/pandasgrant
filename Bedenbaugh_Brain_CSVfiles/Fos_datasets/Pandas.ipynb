```python
weather_df = pd.DataFrame(np.random.rand(20,2)*5,
                          index=pd.date_range(start="2021-01-01", periods=20),
                          columns=["Tokyo", "Beijing"])

def rain_condition(v):
    if v < 1.75:
        return "Dry"
    elif v < 2.75:
        return "Rain"
    return "Heavy Rain"

def make_pretty(styler):
    styler.set_caption("Weather Conditions")
    styler.format(rain_condition)
    styler.format_index(lambda v: v.strftime("%A"))
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
    return styler

weather_df

def make_prettier(styler):
    styler.set_caption("Weather Conditions")
    styler.format(rain_condition)
    styler.format_index(lambda v: v.strftime("%A"))
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="twilight_shifted")
    return styler
```

```python
from allensdk.api.queries.ontologies_api import OntologiesApi
from allensdk.core.structure_tree import StructureTree

# Initialize OntologiesApi
oapi = OntologiesApi()
structure_graph = oapi.get_structures_with_sets([1])  # 1 is the id of the adult mouse structure graph

# Clean the structure_graph
structure_graph = StructureTree.clean_structures(structure_graph)

# Function to create a dictionary mapping structure ids to their respective RGB triplets
def get_colormap(structure_graph):
    colormap = {}
    for structure in structure_graph:
        colormap[structure['id']] = structure['rgb_triplet']
    return colormap

# Get the colormap for all regions
colormap = get_colormap(structure_graph)

import pandas as pd

# Convert the colormap dictionary to a pandas DataFrame
colormap_df = pd.DataFrame(list(colormap.items()), columns=['Structure ID', 'RGB Triplet'])

# Display the colormap DataFrame
colormap_df.head()

```
